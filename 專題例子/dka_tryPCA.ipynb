```python
import pandas as pd

# 指定檔案路徑
file_path = "Hyperchloremia and DKA Dataset .xlsx"

# 讀取 Excel 並指定正確的標題行
data = pd.read_excel(file_path, header=1)  # 1 表示第二行為標題（索引從 0 開始）


# 查看資料
print(data)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 假設 data 是你的 DataFrame
# 自動從 data 中提取所有特徵列
features = data.columns.tolist()

x = data[features]

# 使用插值法處理遺失值
x = x.interpolate(method='linear', axis=0)

# 如果還有遺失值，使用均值填補
x = x.fillna(x.mean())

# 標準化數據
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# 初始化PCA，設置主成分數量
pca = PCA(n_components=14, random_state=42)  # 根據需要調整主成分數量

# 擬合PCA模型並轉換數據
principal_components = pca.fit_transform(x_scaled)

# 將主成分轉換為DataFrame
principal_df = pd.DataFrame(data=principal_components)

#columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5']
print(principal_df.head())


# 主成分的變數名稱和解釋比例
loadings = pd.DataFrame(pca.components_.T, index=features)
explained_variance_ratio = pca.explained_variance_ratio_

print("主成分的變數貢獻度:")
print(loadings)
print("\n解釋比例:")
print(explained_variance_ratio)
print(sum(explained_variance_ratio))

print(principal_df.head())
```
